```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("./src/IBM_HR_ATTRITION.csv")
data.head()
```

```python
data.columns
```

```python
data.Attrition.unique()
# Cette méthode unique() retourne une liste contenant les valeurs uniques de la colonne Attrition
# Ainsi je vérifie si on a bien que des Yes et des No
# Note : La liste retournée n'est pas triée
```

```python
data2 = data.copy()
```

```python
data2.Attrition.value_counts()
# Ceci permet de voir quelles valeurs on a 
# On voit qu'on a soit des 0 soit des 1
```

```python
data2['Attrition'] = data2['Attrition'].apply(lambda x: 1 if x.lower()=="yes" else 0)
# Je remplace la value de l'attrition par des valeurs numériques : 1 ou 0 pour Yes ou No
```

```python
sns.lmplot(x="DistanceFromHome", y="Attrition", data = data2, logistic = True, scatter_kws={"alpha":.05})
# scatter_kws permet de mettre une épaisseur autour de la droite
# logistic = True permet d'avoir une courbe et non une droite toute droite
```

```python
sns.lmplot(x="YearsWithCurrManager", y="Attrition", data = data2, logistic = True, scatter_kws={"alpha":.05})
```

```python
sns.lmplot(x="YearsAtCompany", y="Attrition", data = data2, logistic=True, scatter_kws={"alpha":.05})
```
