```python
import pandas as pd
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
# ceci permet de cacher les warnings que seaborn affiche
```

```python
data = pd.read_csv("./src/IBM_HR_ATTRITION.csv")
data.head()
```

```python
sns.distplot(data.MonthlyIncome, bins = 40)
# KDE kernel density estimate
# bins permet de définir la weight de la barre ou le nombre de barres (?)
```

```python
data2 = data.copy()
data2 = data2[data2["MonthlyIncome"] < 8000]
sns.distplot(data2.MonthlyIncome, bins = 40)
```

```python
sns.distplot(data.loc[data['MonthlyIncome'] < 8000, 'MonthlyIncome'], bins = 40)
# On peut appliquer le mask de cette manière
```
